```python
import matplotlib.pyplot as plt
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
# Создание уравнения
x = Symbol('x')
f = 5*x*x + 10*x -30
f
```

```python
# нахождение корней
a = solve(f,x)
a
```

```python
# вывод на печать с корнями
plot(f, markers=[{'args': [a[1], 0, 'go']}, {'args': [a[0], 0, 'go']}])
```
